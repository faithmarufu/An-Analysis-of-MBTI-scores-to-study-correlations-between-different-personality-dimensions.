## Description
This synthetic dataset is designed to explore and predict Myers-Briggs Type Indicator (MBTI) personality types based on a combination of demographic factors, interest areas, and personality scores. It includes 100K+ samples, each representing an individual with various features that contribute to determining their MBTI type. The dataset can be used to study correlations between different personality dimensions and external factors such as age, gender, education, and interests

# Feature Descriptions

Age: A continuous variable representing the age of the individual.

Gender: A categorical variable indicating the gender of the individual. Possible values are 'Male' and 'Female'.

Education: A binary variable, A value of 1 indicates the individual has at least a graduate-level education (or higher), and 0 indicates an undergraduate, high school level or Uneducated.

Interest: A categorical variable representing the individual's primary area of interest.

Introversion Score: A continuous variable ranging from 0 to 10, representing the individual's tendency toward introversion versus extraversion. Higher scores indicate a greater tendency toward extraversion.

Sensing Score: A continuous variable ranging from 0 to 10, representing the individual's preference for sensing versus intuition. Higher scores indicate a preference for sensing.

Thinking Score: A continuous variable ranging from 0 to 10, indicating the individual's preference for thinking versus feeling. Higher scores indicate a preference for thinking.

Judging Score: A continuous variable ranging from 0 to 10, representing the individual's preference for judging versus perceiving. Higher scores indicate a preference for judging.

Personality: Target that contains People Personality Typeity Type

## Checking for Outliers and Replacing them with median

```python
#Fig1
#Plot Age
sns.histplot(data=df, x='Age')
```

```python
#define q1,q3, iqr
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
iqr=q3-q1
```

```python
# Define the lower bound 
lower_bound=q1-1.5*iqr
```

```python
# Define the upper bound 
upper_bound=q3+1.5*iqr
```

```python
# Find the median of the column
median=df['Age'].median()
```

```python
# Count outliers
df[(df['Age']<lower_bound)|(df['Age']>upper_bound)].count()
```

```python
# Replace values outside the bounds with the median
df['Age']=np.where((df['Age']<lower_bound)|(df['Age']>upper_bound), median, df['Age'])
```

```python
# Count outliers after replacing with median
df[(df['Age']<lower_bound)|(df['Age']>upper_bound)].count()
```

```python
# Fig 2
# check for outliners in age after replacing with median using histplot
sns.histplot(data=df,x='Age')
```

```python
# Categorizing personality types
def categorize_personality(personality):
    if personality in ['ISTP', 'ISFP', 'ESTP', 'ESFP']:
        return 'Explorer'
    elif personality in ['INFJ', 'INFP', 'ENFJ', 'ENFP']:
        return 'Diplomat'
    elif personality in ['INTJ', 'INTP', 'ENTJ', 'ENTP']:
        return 'Analyst'
    elif personality in ['ISTJ', 'ISFJ', 'ESTJ', 'ESFJ']:
        return 'Sentinel'
```

```python
# Create a new column Personality Category
df['Personality_Category'] = df['Personality'].apply(categorize_personality)
df.head(2)
```

## Decision Tree: Hyperparameter Tuned

```python
# For hyperparameter tuning # import GridSearchCV from sklearn.model_selection 
from sklearn.model_selection import GridSearchCV
```

```python
# Grid of parameters to choose from
dtc = DecisionTreeClassifier()
weights = [{0:0.27 ,1:0.73},'balanced',{0:0.5, 1:0.5}]
parameters ={'max_depth':[2,3,4,5,6,7,8,9],
             'min_samples_leaf':[2,5,10,20,25], 'criterion': ['gini','enropy'], 'class_weight': weights}
```

```python
# import metrics from sklearn
from sklearn import metrics
# Setting type of scoring (metrics.make_scorer) used to compare parameter combinations-f1_score for class 1 (pos_label)
scorer=metrics.make_scorer(f1_score, pos_label=1)
```

```python
# set the GridSearchCV with estimator, param_grid, scoring, cv=5
gridCV=GridSearchCV(dtc, parameters, scoring=scorer, cv=5)
```

```python
# fit gridCV on train data; # Set the classifier to the best combination of parameters(best_estimator_); save it as dtc_tuned
dtc_tuned=gridCV.fit(X_train, y_train).best_estimator_
dtc_tuned
```

```python
# predict y (pred_dtc2) on X_test 
pred_dtc2=dtc_tuned.predict(X_test)
```

```python
#Fig 16
# print classification_report 
print(classification_report(y_test, pred_dtc2))
# display ConfusionMatrixDisplay 
ConfusionMatrixDisplay.from_predictions(y_test, pred_dtc2, cmap='Blues')
```

## Choosing the best model

```python
print(y_test.shape, pred_dtc.shape, pred_dtc2.shape, pred_knc.shape)
```

```python
from sklearn.metrics import f1_score
# Ensure data shapes are correct
assert len(y_test) == len(pred_dtc)
assert len(y_test) == len(pred_dtc2)
assert len(y_test) == len(pred_knc)
```

```python
# Calculate the f1 score (an indicator of the model performance) for each model
f1_dtc = f1_score(y_test, pred_dtc, average='weighted')
f1_dtc2 = f1_score(y_test, pred_dtc2, average='weighted')
f1_knc = f1_score(y_test, pred_knc, average='weighted')
```

```python
print(f"F1 score for DTC: {f1_dtc}")
print(f"F1 score for DTC2: {f1_dtc2}")
print(f"F1 score for KNC: {f1_knc}")
```

```python
# set the best_model 
best_model=max((f1_dtc,'Decision Tree Classifier'),(f1_dtc2,'Decision Tree Classifier Hypterparameter Tuned'),(f1_knc,'KNeighbors Classifier'))
```

```python
# print the best model
print(f'The best model is {best_model[1]} with f1_socre of {best_model[0]}')
```

```python
from sklearn.metrics import recall_score

# Using 'macro' average for multiclass recall
recall_dtc = recall_score(y_test, pred_dtc, average='weighted')
recall_dtc2 = recall_score(y_test, pred_dtc2, average='weighted')
recall_knc = recall_score(y_test, pred_knc, average='weighted')
```

```python
print(f"Recall (DTC): {recall_dtc}")
print(f"Recall (DTC2): {recall_dtc2}")
print(f"Recall (KNC): {recall_knc}")
```

```python
# set the best_model 
best_model2=max((recall_dtc,'Decision Tree Classifier'),(recall_dtc2,'Decision Tree Classifier Hyperparameter Tuned'),(recall_knc,'KNeighbors Classifier'))
```

```python
# print the best model 
print(f'The best model is {best_model2[1]} with recall_socre of {best_model2[0]}')
```
